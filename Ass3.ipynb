```python
Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.

Note: timepoint, signal, event, and region are columns in the fmri dataset.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fmri=sns.load_dataset('fmri')
sns.lineplot(data=fmri, x='timepoint', y='signal',hue='event',style='region')

plt.title("FMRI Signal Over Time by Event and Region")
plt.xlabel("Time point")
plt.ylabel("Signal")
plt.legend(title='Event and Region')
plt.show()
```

```python
Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.

Note: pclass, age, and fare are columns in the titanic dataset.
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
titanic =sns.load_dataset("titanic")
fig,axes=plt.subplots(1,2,figsize=(16,6))
sns.boxplot(ax=axes[0], data=titanic,x='pclass',y='age',palette='Set2')
axes[0].set_tittle('Box Plot of Age by Passenger Class')
axes[0].set_xlabel('Passenger Class')
axes[0].set_ylabel('Age')

sns.boxplot(ax=axes[1], data=titanic, x='pclass',y='fare',palette='Set2')
axes[1].set_tittle('Box Plot of Age by Passenger Class')
axes[1].set_xlabel('Passenger Class')
axes[1].set_ylabel('Fare')

plt.tight_layout()
plt.show()
```

```python
Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
diamonds=sns.load_dataset("diamonds")
plt.figure(figsize=(12,6))
sns.histplot(data=diamonds, x='price',hue='cut',multiple='stack', palette='tab10',edgecolor=".3",bins=50)
plt.tittle('Distribution of Diamond Prices by Cut')
plt.xlabel('Price')
plt.ylabel('Count')

plt.show()
```
